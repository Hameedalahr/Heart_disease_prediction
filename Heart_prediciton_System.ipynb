Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection & Processing

```python
#loading the csv data into Panda's Dataframes
heart_data=pd.read_csv('/content/heart.csv')
```

```python
#printing first 5 rows of the dataset
heart_data.head()
```

```python
#printing the last 5 rows of the dataset
heart_data.tail()
```

```python
#finding the number of rows and columns in the dataset
heart_data.shape
```

```python
#getting some info about the data
heart_data.info()
```

```python
#checking for missing values
heart_data.isnull().sum()
```

```python
#statstical measures about the data
heart_data.describe()
```

```python
#checking the distribution of Target variable
heart_data['target'].value_counts()
```

1-->Defective heart
0-->healthy heart

```python
X=heart_data.drop(columns='target',axis=1)
Y=heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into Training data & Test data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(Y.shape,Y_train.shape,Y_test.shape)
```

**Model Training**

Logistic Regression

```python
model=LogisticRegression()
```

```python
#training the LogisticRegression model with Training data
model.fit(X_train,Y_train)
```

Model Evaluation

Accuracy Score

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on Training data: ',training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on Tesdt data:",test_data_accuracy)
```

Building a Predictive System

```python
import pandas as pd

# Assuming these are the feature names used during training
feature_names = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']

input_data = (71, 0, 0, 112, 149, 0, 1, 125, 0, 1.6, 1, 0, 2)
input_data_as_numpy_array = np.asarray(input_data)
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# Create a DataFrame with feature names
input_df = pd.DataFrame(input_data_reshaped, columns=feature_names)

prediction = model.predict(input_df)
print(prediction)
if prediction == 0:
    print("The person doesn't have any Heart Disease")
else:
    print("The person has Heart Disease")
```
